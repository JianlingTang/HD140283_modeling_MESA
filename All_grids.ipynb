```python
def fastExpMod(b, e):
    result = 1
    print(e)
    print(e%2)
    while e != 0:
        if e%2 == 1:
            # ei = 1, then mul
            result = (result * b)
        e = e / 2
        print(e/2)
        # b, b^2, b^4, b^8, ... , b^(2^n)
        b = (b*b)
    return result
fastExpMod(2,3)
```
